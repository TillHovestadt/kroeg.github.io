```python
import pandas as pd
import hvplot.pandas
import panel as pn
import geopandas as gpd
from shapely.geometry import Point
import geoviews
```

```python
# load data 

dummy_data = gpd.read_file("../data/synthetic_data.gpkg")
gdf = pd.DataFrame(dummy_data).copy()
gdf = gdf.reset_index(drop=True)
```

```python
gdf = gdf.reset_index(drop=True)

vars_ = ["kroeg_index1", "kroeg_index2", "kroeg_index3", "kroeg_index4", "kroeg_index5"]

plots = {}

for var in vars_:

    plots[var] = gdf.hvplot.polygons(
        c=var,
        geo=True,
        tiles="CartoLight",
        cmap="viridis",
        colorbar=True,
        hover_cols=["postcode", var],
        frame_height=700,
    )
```

```python
pn.extension()

vars_ = ["kroeg_index1", "kroeg_index2", "kroeg_index3", "kroeg_index4", "kroeg_index5"]

indicator = pn.widgets.DiscreteSlider(

    name="Indicator",
    options=vars_,
    value=vars_[0]

)

def make_map(var):

    vmin = gdf[var].min()
    vmax = gdf[var].max()

    return gdf.hvplot.polygons(
        c=var,
        cmap="viridis",
        clabel=var,
        colorbar=True,
        geo=True,
        tiles=True,
        data_aspect=1,
        frame_height=700,
        frame_width=900,
        hover_cols=["postcode", var],
        title=f"Postal code map: {var}",
    )#.opts(clim=(vmin, vmax))

dashboard = pn.Column(

    indicator,

    pn.bind(make_map, indicator)

)

dashboard.servable()
```
