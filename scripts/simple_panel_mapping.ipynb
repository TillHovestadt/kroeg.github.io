```python
import pandas as pd
import hvplot.pandas
import panel as pn
import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon
from matplotlib.cm import Reds, Blues, Greens
import cartopy.crs as ccrs
```

```python
# load data 

the_data = gpd.read_file("../data/indexed_data_polygons.gpkg")
gdf = the_data.copy()
gdf = gdf.set_crs("EPSG:28992", allow_override=True)
gdf = gdf.reset_index(drop=True)
```

```python
gdf["geometry"] = gdf["geometry"].simplify(tolerance=1000, preserve_topology=True)
```

```python
sectors_ = ["third_places_total_ranked", 
            #"third_places_non_medical_total_ranked",
            #"third_places_non_medical_non_retail_total_ranked",
            "third_places_medical_ranked",
            "third_places_food_beverage_ranked",
            "third_places_retail_ranked",
            "third_places_community_ranked",
            "third_places_cultural_ranked",
            "third_places_sport_leisure_ranked"]

LABELS = {
    "All Third Places": "third_places_total_ranked",
    #"All Third Places except Medical Places": "third_places_non_medical_total_ranked",
    #"All Third Places except Medical and Retail Places": "third_places_non_medical_non_retail_total_ranked",
    "Medical Places only": "third_places_medical_ranked",
    #"Food and Beverage only": "third_places_food_beverage_ranked",
    #"Retail Places only": "third_places_retail_ranked",
    "Community Places only": "third_places_community_ranked",
    #"Cultural Places only": "third_places_cultural_ranked",
    #"Sport and Leisure only": "third_places_sport_leisure_ranked",
}
```

```python
pn.extension(design="material", sizing_mode="stretch_width")
```

```python
def create_map(sector):
    return gdf.hvplot.polygons(
        c=sector,
        geo=True,
        crs=ccrs.epsg(28992),
        projection=ccrs.GOOGLE_MERCATOR,
        tiles="CartoLight",
        cmap="Reds",
        colorbar=True,
        hover_cols=["postcode", sector],
        frame_width=900,
        frame_height=700,
    )
```

```python
create_map("third_places_total_ranked")
```

```python
sector_widget = pn.widgets.Select(label="Sector", 
                                  options=dict(LABELS))
```

```python
bound_plot = pn.bind(
    create_map, sector=sector_widget
)
```

```python
#widgets = pn.Column(sector_widget, sizing_mode="fixed", width=300)
#pn.Column(widgets, bound_plot)#.show()
```

```python
pn.template.MaterialTemplate(
    site="Panel",
    title="Getting Started App",
    sidebar=[sector_widget],
    main=[bound_plot],
).servable(); # The ; is needed in the notebook to not display the template. Its not needed in a script
```
